```python
import pandas as pd
```

```python
from zipfile import ZipFile
from urllib.request import urlopen
from io import BytesIO

with urlopen("https://www2.census.gov/topics/genealogy/2010surnames/names.zip") as f_zip:
    f = ZipFile(BytesIO(f_zip.read()))

df = pd.read_csv(f.open("Names_2010Census.csv"))
df
```

```python
df_last_name = df[df["name"] != "ALL OTHER NAMES"]
df_last_name
```

```python
df_last_name["name"] = df_last_name["name"].str.capitalize()
df_last_name
```

```python
df_last_name[["name", "count"]].to_csv("last-name.csv", index=False)
```
