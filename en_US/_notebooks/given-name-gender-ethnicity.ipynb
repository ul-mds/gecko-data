```python
import pandas as pd
```

```python
df = pd.read_csv("https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv?accessType=DOWNLOAD")
df
```

```python
df_given_name = df.rename(columns={
    "Gender": "gender",
    "Child's First Name": "given_name",
    "Count": "count",
    "Ethnicity": "ethnicity",
})[["gender", "given_name", "count", "ethnicity"]]
df_given_name
```

```python
# M = male, F = female
df_given_name["gender"].unique()
```

```python
# H = hispanic, W = white non hisp(anic), A = asian and paci(fic islander), B = black non hisp(anic)
df_given_name["ethnicity"].unique()
```

```python
# remap gender into gender code
df_given_name["gender_code"] = df_given_name["gender"].map({
    "FEMALE": "f",
    "MALE": "m",
})
```

```python
# remap ethnicity into ethnicity code
df_given_name["ethnicity_code"] = df_given_name["ethnicity"].map({
    "HISPANIC": "h",
    "WHITE NON HISP": "w",
    "WHITE NON HISPANIC": "w",
    "ASIAN AND PACI": "a",
    "ASIAN AND PACIFIC ISLANDER": "a",
    "BLACK NON HISP": "b",
    "BLACK NON HISPANIC": "b"
})
```

```python
# capitalize strings
df_given_name["given_name"] = df_given_name["given_name"].str.capitalize()
```

```python
df_slim = df_given_name[["given_name", "gender_code", "ethnicity_code", "count"]]
df_slim
```

```python
df_slim.groupby(["given_name", "gender_code", "ethnicity_code"]).sum().to_csv("given-name-gender-ethnicity.csv")
```
